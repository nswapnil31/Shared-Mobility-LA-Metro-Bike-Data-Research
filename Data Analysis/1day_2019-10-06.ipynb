```python
# Import Libraries

import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#read CSV File

Year_2019 = pd.read_csv("D:\\GA Work_Essia\\Data\\metro-bike-share-trips-2019-q4.csv")
Year_2019.head()
```

```python
#Change datatype
    
Year_2019["start_time"]=pd.to_datetime(Year_2019['start_time'])
Year_2019["end_time"]=pd.to_datetime(Year_2019['end_time'])
```

```python
#extract date from time

Year_2019["start_day"]=Year_2019["start_time"].dt.date
Year_2019["end_day"]=Year_2019["end_time"].dt.date

#extract hour from time

Year_2019["start_hour"]=Year_2019["start_time"].dt.hour
Year_2019["end_hour"]=Year_2019["end_time"].dt.hour
# Find day name

Year_2019["weekday"]=Year_2019["start_time"].dt.weekday_name

```

```python
Year_2019_Q4=Year_2019[['start_station','start_day','start_hour','weekday','end_station','end_day','end_hour','duration','bike_id','passholder_type','plan_duration','trip_route_category','bike_type']]
Year_2019_Q4.head()
```

```python
Year_2019_Q4['start_day'].value_counts().head(1)
```

```python
Year_2019_Q4["start_day"]=pd.to_datetime(Year_2019_Q4['start_day'])
```

```python
data_1day=Year_2019_Q4[(Year_2019_Q4['start_day'] > '2019-10-05') & (Year_2019_Q4['start_day'] < '2019-10-07')]
data_1day=data_1day[(data_1day['start_hour'] > 7) & (data_1day['start_hour'] < 21)]
```

```python
data_1day['start_station'].value_counts().head()
```

### Smart Bike available only in few stations and no sttaion has all three bikes together

```python
data_3005_start=data_1day[(data_1day['start_station'] > 3004) & (data_1day['start_station'] < 3006)]
data_3005_end=data_1day[(data_1day['end_station'] > 3004) & (data_1day['end_station'] < 3006)]
```

```python
data_3005_start["bike_Left_count"] = data_3005_start["bike_id"].count()
data_3005_end["bike_Arrive_count"] = data_3005_end["bike_id"].count()
```

```python
Left=pd.pivot_table(data_3005_start,index=["start_station","start_hour"],columns=["bike_type"],values=["bike_Left_count"],aggfunc="count",margins=True)
Left
```

```python
Left.columns = ["_".join((j,k)) for j,k in Left.columns]
Left.reset_index()
```

```python
Arrive=pd.pivot_table(data_3005_end,index=["end_station","end_hour"],columns=["bike_type"],values=["bike_Arrive_count"],aggfunc="count",margins=True)
print(Arrive)
```

```python
Arrive.columns = ["_".join((j,k)) for j,k in Arrive.columns]
Arrive.reset_index()
```

```python
result=pd.concat([Left,Arrive],axis=1)
result
result.reset_index()
```
