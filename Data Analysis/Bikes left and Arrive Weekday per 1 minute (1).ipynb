```python
# Import Libraries

import pandas as pd
import os
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

no_color = sns.color_palette()[0]

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# import file
data=pd.read_csv('D:\\GA Work_Essia\\Data\\Data Files\\Trip Data\\2019\\All_2019_data.csv')
data.head()
```

```python
# Changing the format to datetime
data.start_time = pd.to_datetime(data.start_time)
data.end_time = pd.to_datetime(data.end_time)
```

```python
# Changing the trip_id, start_station_id, and end_station_id to object
data.bike_id = data.bike_id.astype(str)
data.trip_id = data.trip_id.astype(str)
data.start_station = data.start_station.astype(str)
data.end_station = data.end_station.astype(str)
```

```python
# Changing the trip_r to category format
data.trip_route_category = data.trip_route_category.astype('category')
data.passholder_type = data.passholder_type.astype('category')
data.bike_type = data.bike_type.astype('category')
```

```python
# Start time month (January - December)
data['start_time_month'] = data['start_time'].dt.strftime('%B')
```

```python
# Start time weekday
data['start_time_weekday'] = data['start_time'].dt.strftime('%a')
```

```python
#extract date from time

data["start_day"]=data["start_time"].dt.date
data["end_day"]=data["end_time"].dt.date
```

```python
# Week Number
data['Week_Number'] = data['start_time'].dt.week
```

```python
# Start and end time hour
data['start_time_hour'] = data['start_time'].dt.hour
data['end_time_hour'] = data['end_time'].dt.hour
# Start and end time hour
data['start_time_min'] = data['start_time'].dt.minute
data['end_time_min'] = data['end_time'].dt.minute
```

```python
data['duration'] = data['duration'].astype(int)
```

```python
data['new_Start_time'] = [d.time() for d in data['start_time']]
data['new_end_time'] = [d.time() for d in data['end_time']]
```

```python
data.drop(['start_time','end_time','start_lat','start_lon','end_lat','end_lon'], axis=1, inplace=True)
```

```python
pd.set_option('display.max_columns', None)
data.head()
```

```python
# filering data for station 3005
data_3005_start=data.query('start_station =="3005"')
data_3005_end=data.query('end_station=="3005"')
```

```python
#count of no of bike left and arrive
data_3005_start["Left"] = data_3005_start["bike_id"].count()
data_3005_end["Arrive"] = data_3005_end["bike_id"].count()
```

```python
#Average Number of bikes left every day from station 3005 per minuate in 1 year (Considering 52 weeks in year to find average of each day I have divided by 52)

pd.set_option('display.max_rows', None)

Bikes_Left_per_Min=pd.pivot_table(data_3005_start,index=["start_time_hour","start_time_min"],columns=["start_time_weekday"],values=["Left"],aggfunc="count",margins=True)/52
Bikes_Left_per_Min.columns = ["_".join((j,k)) for j,k in Bikes_Left_per_Min.columns]
Bikes_Left_per_Min.reset_index()
```

```python
#Average Number of bikes arrive every day from station 3005 per minuate in 1 year (Considering 52 weeks in year to find average of each day I have divided by 52)


pd.set_option('display.max_rows', None)
Bikes_Arrive_per_Min=pd.pivot_table(data_3005_end,index=["start_time_hour","end_time_min"],columns=["start_time_weekday"],values=["Arrive"],aggfunc="count",margins=True)/52
Bikes_Arrive_per_Min.columns = ["_".join((j,k)) for j,k in Bikes_Arrive_per_Min.columns]
Bikes_Arrive_per_Min.reset_index()
```

```python
#Average Number of bikes left and arrive every day from station 3005 per minuate in 1 year (Considering 52 weeks in year to find average of each day I have divided by 52)

Bike_per_min=pd.concat([Bikes_Left_per_Min,Bikes_Arrive_per_Min],axis=1)
Bike_per_min

Bike_per_min.reset_index()
Bike_per_min = pd.DataFrame(Bike_per_min.to_records())

```

```python
Bike_per_min.rename(columns={'level_0': 'Hour', 'level_1': 'Min'})
```

```python
# Average Number of bikes left every day from station 3005 per hour in 1 year 
Bikes_Left_per_Hour=pd.pivot_table(data_3005_start,index=["start_time_hour"],columns=["start_time_weekday"],values=["Left"],aggfunc="count",margins=True)/52
Bikes_Left_per_Hour.columns = ["_".join((j,k)) for j,k in Bikes_Left_per_Hour.columns]
Bikes_Left_per_Hour.reset_index()
```

```python
# Average Number of bikes arrive every day from station 3005 per hour in 1 year 
Bikes_Arrive_per_Hour=pd.pivot_table(data_3005_end,index=["end_time_hour"],columns=["start_time_weekday"],values=["Arrive"],aggfunc="count",margins=True)/52
Bikes_Arrive_per_Hour.columns = ["_".join((j,k)) for j,k in Bikes_Arrive_per_Hour.columns]
Bikes_Arrive_per_Hour.reset_index()
```

```python
# Average Number of bikes left and arrive every day from station 3005 per hour in 1 year 
Bike_per_Hour=pd.concat([Bikes_Left_per_Hour,Bikes_Arrive_per_Hour],axis=1)
Bike_per_Hour
Bike_per_Hour.reset_index()
```
